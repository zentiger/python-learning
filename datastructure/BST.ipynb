```python
# AVL-Tree
class TreeNode():

    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.height = 0
        self.parent = None

class AVLTree():
    """AVL Tree: A balanced BST, created by Adelson-Velskii and Landis."""

    def __init__(self, root = None):
        self.root = root

    def insert(self, val):
        if not self.root:
            return TreeNode(val)
        pass 
    
    def search(self, val):
        if not self.root:
            self.insert(val)
        curr = self.root
        while curr:
            if curr.val == val:
                return curr
            elif curr.val > val:
                curr = curr.left
            elif curr.val < val:
                curr = curr.right
        else:
            self.insert(val)

    def remove(self, val):
        pass

    def _is_balanced(self, node):
        if node.left and node.right and node.left.height == node.right.height:
            return True
        return False
    
    def _balance_factor(self, node):
        if node.left and node.right:
            return node.left.height - node.right.height
        if node.left:
            return 1
        if node.right:
            return -1 
        


```
